```python
# OYUN KURALLARI
# SİZE 0 İLE 10 ARASINDA 3 DEFA KART ÇEKME HAKKI TANINACAKTIR.
# VERİLEN SAYIYI EN YAKIN 21'E YAKLAŞTIRAN KAZANIR
# EĞER 21 GEÇERSENİZ KAYBEDERSİNİZ, EĞER RAKİBİNİZ VE SİZ 21 GEÇERSENİZ KAYBEDERSİNİZ.
# HERHANGİ Bİ' BERABERE KALMA DURUMUNDA KAYBEDEN SİZ OLURSUNUZ.
# SİZE VERİLEN SAYIYI ARTIRMAK VEYA OLDUĞU GİBİ BIRAKMAK SİZİN SEÇİMİNİZ
# OYUNA BAŞLADIĞINIZDA 100 KREDİNİZ BULUNMAKTADIR. HER OYUNA BAŞLADIĞINIZDA KAÇ KREDİ İLE GİRMEK İSTEDİĞİNİZİ YAZIYORSUNUZ,
# KAZANIRSANIZ GİRDİĞİNİZ BAHİS KADAR KAZANIYOR, KAYBEDERSENİZ GİRDİĞİNİZ KREDİ BAHİS KADAR KAYBETMİŞ OLUYORSUNUZ.
from tkinter import *
from tkinter import messagebox
import random
from PIL import Image, ImageTk
from tkinter import Tk, Label, Frame, Canvas
import tkinter as tk
class Kredi():
    def __init__(self,kredim=100):
        self.kredim = kredim
kredi_miktari = Kredi() 
```

```python
#f0f0f0 canvas ana renk
class Yirmi_Bir():
    def __init__(self, master,oyuncunun_skoru = 0,bilgisayarin_skoru = 0,oyun_hakki = 3,oyun_sonu = 0):
        self.master = master
        self.oyuncunun_skoru = oyuncunun_skoru
        self.bilgisayarin_skoru = bilgisayarin_skoru
        self.oyun_hakki = oyun_hakki
        self.oyun_sonu = oyun_sonu
        self.canvas = Canvas(master, height = 450, width = 650)
        self.canvas.pack()

        #RAKİP FRAME + TEXT
        self.rakip_frame = Frame(self.master, bg = "#add8e6")
        self.rakip_frame.place(relheight=0.5,relwidth=0.43, relx=0.05, rely=0.05)
        self.rakip_text = Label(self.rakip_frame,bg="#add8e6",text="RAKİP",font="Verdana 12 bold")
        self.rakip_text.place(relx=0.38,rely=0.05)
        
        #OYUNCU FRAME + TEXT
        self.oyuncu_frame = Frame(self.master, bg = "#add8e6")
        self.oyuncu_frame.place(relheight=0.5,relwidth=0.43, relx=0.52, rely=0.05)
        self.oyuncu_text = Label(self.oyuncu_frame,bg="#add8e6",text="OYUNCU",font="Verdana 12 bold")
        self.oyuncu_text.place(relx=0.35,rely=0.05)
        
        #BASLANGIC + BUTTON
        self.button_frame = Frame(self.master, bg = "#f0f0f0")
        self.button_frame.place(relheight=0.5,relwidth=0.43, relx=0.3, rely=0.55)

        self.baslangic_button = Button(self.button_frame,
                                  text="Oyunu başlatmak için bana tıkla!!",
                                  command=self.oyunu_baslat,
                                  width=28,
                                  height=2,
                                  font="Verdana 11",
                                  bg="#ffffff",
                                  wraplength=300)
        self.baslangic_button.place(rely=0.2)
        self.baslangic_rakip_text = Label(self.rakip_frame,bg="#add8e6",text="Oyun başlamayı bekliyor...",font="Verdana 12")
        self.baslangic_rakip_text.place(relx=0.088,rely=0.45)
        self.baslangic_oyuncu_text = Label(self.oyuncu_frame,bg="#add8e6",text="Oyun başlamayı bekliyor...",font="Verdana 12")
        self.baslangic_oyuncu_text.place(relx=0.088,rely=0.45)
        self.value = None 
        
    def oyunu_baslat(self):
        self.baslangic_button.destroy()
        self.baslangic_rakip_text.destroy()
        self.baslangic_oyuncu_text.destroy()
        messagebox.showinfo("Başarılı işlem","Öncelikle bahis işlemlerini halledelim.")
    
        self.bahis_text = Label(self.button_frame,bg="#f0f0f0",text="Bahis miktarını giriniz: ",font="Verdana 12")
        self.bahis_text.place(relx=0,rely=0.20)
        self.validation = self.master.register(self.sayi_girisi)
        self.bahis_alani = Entry(self.button_frame, validate="key", validatecommand=(self.validation, "%P"))
        self.bahis_alani.bind("<KeyRelease>", self.bahis_yazisi_guncelleme)
        self.bahis_alani.place(relx=0.7,rely=0.207)

        self.bahis_onay_button = Button(self.button_frame,
                                        text="Bahisi Onayla",
                                        command=self.bahis_girisi,
                                        width=15,
                                        height=2,
                                        font="Verdana 10",
                                        bg="#ffffff")
        self.bahis_onay_button.place(relx=0.25,rely=0.4)

        self.girilen_bahis_text = Label(self.master,bg="#f0f0f0",text="Girdiğiniz bahis miktarı: ",font="Verdana 8")
        self.girilen_bahis_text.place(relx=0.05,rely=0.87)

        self.label_text = StringVar() 
        self.girilen_bahis_gorunmesi_text = Label(self.master,
                                                  bg="#f0f0f0",
                                                  text="0 ",
                                                  font="Verdana 8",
                                                  textvariable=self.label_text)
        self.girilen_bahis_gorunmesi_text.place(relx=0.26,rely=0.871)
        
        self.mevcut_kredi_text = Label(self.master,bg="#f0f0f0",text="Sahip olduğunuz kredi miktarı: ",font="Verdana 8")
        self.mevcut_kredi_text.place(relx=0.05,rely=0.915)

        self.mevcut_kredi_miktari_text = Label(self.master,bg="#f0f0f0",text=kredi_miktari.kredim,font="Verdana 8")
        self.mevcut_kredi_miktari_text.place(relx=0.323,rely=0.915)
        
    
    def sayi_girisi(self,V):
        if V == "" or V.isdigit():
            return True
        else:
            messagebox.showerror("Hata", "Sadece sayı girebilirsiniz.(bahis yapmak istemiyorsanız 0 yazın)")
            return False
        
    def bahis_yazisi_guncelleme(self,*args):
        entered_text = self.bahis_alani.get()
        self.label_text.set(entered_text)
        self.value = entered_text
     
    def kaybetmek(self):
        if self.oyun_hakki == 0:
            self.rakip_skor_text = Label(self.rakip_frame, bg="#add8e6", text=self.bilgisayarin_skoru, font="Verdana 12 bold")
            self.rakip_skor_text.place(relx=0.47, rely=0.7)
            self.oyuncu_skor_text = Label(self.oyuncu_frame, bg="#add8e6", text=self.oyuncunun_skoru, font="Verdana 12 bold")
            self.oyuncu_skor_text.place(relx=0.47, rely=0.7)

            kredi_miktari.kredim -= self.bahis_miktari
            messagebox.showinfo("Kaybettiniz","Oyunu kaybettiniz kalan kredi miktarınız: {}".format(kredi_miktari.kredim))
            if self.master:
                self.master.destroy()
        else:
            self.resim_goruntule_bilgisayar()
            self.resim_goruntule_oyuncu()
            self.rakip_skor_text = Label(self.rakip_frame, bg="#add8e6", text=self.bilgisayarin_skoru, font="Verdana 12 bold")
            self.rakip_skor_text.place(relx=0.47, rely=0.7)
            self.oyuncu_skor_text = Label(self.oyuncu_frame, bg="#add8e6", text=self.oyuncunun_skoru, font="Verdana 12 bold")
            self.oyuncu_skor_text.place(relx=0.47, rely=0.7)

            kredi_miktari.kredim -= self.bahis_miktari
            messagebox.showinfo("Kaybettiniz","Oyunu kaybettiniz kalan kredi miktarınız: {}".format(kredi_miktari.kredim))
            if self.master:
                self.master.destroy()
    def kazanmak(self):
        if self.oyun_hakki == 0:
            self.rakip_skor_text = Label(self.rakip_frame, bg="#add8e6", text=self.bilgisayarin_skoru, font="Verdana 12 bold")
            self.rakip_skor_text.place(relx=0.47, rely=0.7)
            self.oyuncu_skor_text = Label(self.oyuncu_frame, bg="#add8e6", text=self.oyuncunun_skoru, font="Verdana 12 bold")
            self.oyuncu_skor_text.place(relx=0.47, rely=0.7)

            bahis = self.bahis_miktari * 1
            kredi_miktari.kredim += bahis
            messagebox.showinfo("Kazandınız!!","Oyunu kazandınız kalan kredi miktarınız: {}".format(kredi_miktari.kredim))
            if self.master:
                self.master.destroy()
        else:
            self.resim_goruntule_bilgisayar()
            self.resim_goruntule_oyuncu()
            self.rakip_skor_text = Label(self.rakip_frame, bg="#add8e6", text=self.bilgisayarin_skoru, font="Verdana 12 bold")
            self.rakip_skor_text.place(relx=0.47, rely=0.7)
            self.oyuncu_skor_text = Label(self.oyuncu_frame, bg="#add8e6", text=self.oyuncunun_skoru, font="Verdana 12 bold")
            self.oyuncu_skor_text.place(relx=0.47, rely=0.7)

            bahis = self.bahis_miktari * 1
            kredi_miktari.kredim += bahis
            messagebox.showinfo("Kazandınız!!","Oyunu kazandınız kalan kredi miktarınız: {}".format(kredi_miktari.kredim))
            if self.master:
                self.master.destroy()
    
    
    def son_durum(self):
        self.rakip_skor_text = Label(self.rakip_frame, bg="#add8e6", text=self.bilgisayarin_skoru, font="Verdana 12 bold")
        self.rakip_skor_text.place(relx=0.47, rely=0.7)
        self.oyuncu_skor_text = Label(self.oyuncu_frame, bg="#add8e6", text=self.oyuncunun_skoru, font="Verdana 12 bold")
        self.oyuncu_skor_text.place(relx=0.47, rely=0.7)
        if self.bilgisayarin_skoru == 21 and self.oyuncunun_skoru > 21:
            self.kaybetmek()
        elif self.bilgisayarin_skoru == 21 and self.oyuncunun_skoru < 21:
            self.kaybetmek()
        elif self.oyuncunun_skoru == 21 and self.bilgisayarin_skoru > 21:
            self.kazanmak()
        elif self.oyuncunun_skoru == 21 and self.bilgisayarin_skoru < 21:
            self.kazanmak()
        elif self.oyuncunun_skoru == 21 and self.bilgisayarin_skoru == 21:
            self.kaybetmek()
        elif self.oyuncunun_skoru > 21 and self.bilgisayarin_skoru > 21:
            self.kaybetmek()
        elif self.oyuncunun_skoru < 21 and self.bilgisayarin_skoru < 21:
            if self.oyuncunun_skoru < self.bilgisayarin_skoru:
                self.kaybetmek()
            elif self.oyuncunun_skoru > self.bilgisayarin_skoru:
                self.kazanmak()
            elif self.oyuncunun_skoru == self.bilgisayarin_skoru:
                self.kaybetmek()
            else:
                None
        elif self.oyuncunun_skoru > 21 and self.bilgisayarin_skoru < 21:
            self.kaybetmek()
        elif self.oyuncunun_skoru < 21 and self.bilgisayarin_skoru > 21:
            self.kaybetmek()
        else:
            None
    def genel_durum(self):
        self.rakip_skor_text = Label(self.rakip_frame, bg="#add8e6", text=self.bilgisayarin_skoru, font="Verdana 12 bold")
        self.rakip_skor_text.place(relx=0.47, rely=0.7)
        self.oyuncu_skor_text = Label(self.oyuncu_frame, bg="#add8e6", text=self.oyuncunun_skoru, font="Verdana 12 bold")
        self.oyuncu_skor_text.place(relx=0.47, rely=0.7)

        if self.bilgisayarin_skoru == 21 and self.oyuncunun_skoru > 21: 
            self.kaybetmek()
        elif self.bilgisayarin_skoru == 21 and self.oyuncunun_skoru < 21:
            self.kaybetmek()
        elif self.oyuncunun_skoru == 21 and self.bilgisayarin_skoru > 21:
            self.kazanmak()
        elif self.oyuncunun_skoru == 21 and self.bilgisayarin_skoru < 21:
            self.kazanmak()
        elif self.oyuncunun_skoru == 21 and self.oyuncunun_skoru == self.bilgisayarin_skoru:
            self.kaybetmek()
        elif self.oyuncunun_skoru > 21:
            self.kaybetmek()
        elif self.bilgisayarin_skoru > 21 and self.oyuncunun_skoru <= 21:
            self.kazanmak()
        else:
            None
    def cek(self): 
        self.kalan_el_sayac += 1
        self.oyun_hakki -= 1
        self.kalan_hak = Label(self.button_frame,bg="#f0f0f0",text=self.oyun_hakki,font="Verdana 12 bold")
        self.kalan_hak.place(relx=0.67,rely=0.155)
        self.oyuncu_random_skor_giris = random.randint(1,10)
        self.oyuncunun_skoru += self.oyuncu_random_skor_giris
        self.resim_goruntule_oyuncu()
        
        if self.oyun_hakki == 0:
            if self.bilgisayarin_skoru <= 17:
                self.bilgisayar_random_skor_giris = random.randint(1,10)
                self.bilgisayarin_skoru += self.bilgisayar_random_skor_giris
                self.resim_goruntule_bilgisayar()
                self.son_durum()
            else:
                self.son_durum()
        else:
            if self.bilgisayarin_skoru <= 17:
                self.bilgisayar_random_skor_giris = random.randint(1,10)
                self.bilgisayarin_skoru += self.bilgisayar_random_skor_giris
                self.resim_goruntule_bilgisayar()
                self.genel_durum()
            else:
                self.genel_durum()
                
                
    def pas(self):
        self.kalan_el_sayac += 1
        self.oyun_hakki -= 1
        self.kalan_hak = Label(self.button_frame,bg="#f0f0f0",text=self.oyun_hakki,font="Verdana 12 bold")
        self.kalan_hak.place(relx=0.67,rely=0.155)
        
        if self.oyun_hakki == 0:
            if self.bilgisayarin_skoru <= 17:
                self.bilgisayar_random_skor_giris = random.randint(1,10)
                self.bilgisayarin_skoru += self.bilgisayar_random_skor_giris
                self.resim_goruntule_bilgisayar()
                self.son_durum()
            else:
                self.son_durum()
        else:
            if self.bilgisayarin_skoru <= 17:
                self.bilgisayar_random_skor_giris = random.randint(1,10)
                self.bilgisayarin_skoru += self.bilgisayar_random_skor_giris
                self.resim_goruntule_bilgisayar()
                self.genel_durum()
            else:
                self.genel_durum()
        
        
    def rotate_and_resize(self,image_path, angle, scale_factor):
        self.original_image = Image.open(image_path)
        self.desired_width = int(self.original_image.width * scale_factor)
        self.desired_height = int(self.original_image.height * scale_factor)
        self.resized_image = self.original_image.resize((self.desired_width, self.desired_height), Image.LANCZOS)

        self.rotated_image = self.resized_image.rotate(angle, expand=True)
        return(self.rotated_image)
    
    
    def resim_goruntule_oyuncu(self):
        for x in range(1,11):
            if self.oyuncu_random_skor_giris == x:
                if self.kalan_el_sayac == 1:
                    self.rotated_image1 = self.rotate_and_resize("k_"+ str(x) + ".png", 0, 0.8)
                    self.photo1 = ImageTk.PhotoImage(self.rotated_image1)
                    self.label1 = tk.Label(self.oyuncu_frame,bg="#add8e6", image=self.photo1, borderwidth=0, highlightthickness=0,height=0,width=0)
                    self.label1.image = self.photo1
                    self.label1.place(x=70, y=75)
                elif self.kalan_el_sayac == 2:
                    self.rotated_image2 = self.rotate_and_resize("s_"+str(x) + ".png", 0, 0.8)
                    self.photo2 = ImageTk.PhotoImage(self.rotated_image2)
                    self.label2 = tk.Label(self.oyuncu_frame,bg="#add8e6", image=self.photo2, borderwidth=0, highlightthickness=0,height=0,width=0)
                    self.label2.image = self.photo2
                    self.label2.place(x=90, y=75)
                elif self.kalan_el_sayac == 3:
                    self.rotated_image3 = self.rotate_and_resize("k"+str(x) + ".png", 0, 0.62)
                    self.photo3 = ImageTk.PhotoImage(self.rotated_image3)
                    self.label3 = tk.Label(self.oyuncu_frame,bg="#add8e6", image=self.photo3, borderwidth=0, highlightthickness=0,height=0,width=0)
                    self.label3.image = self.photo3
                    self.label3.place(x=110, y=75)
                elif self.kalan_el_sayac == 4:
                    self.rotated_image4 = self.rotate_and_resize("s"+str(x) + ".png", 0, 0.55)
                    self.photo4 = ImageTk.PhotoImage(self.rotated_image4)
                    self.label4 = tk.Label(self.oyuncu_frame,bg="#add8e6", image=self.photo4, borderwidth=0, highlightthickness=0,height=0,width=0)
                    self.label4.image = self.photo4
                    self.label4.place(x=130, y=75)
                else:
                    None
        
            
    def resim_goruntule_bilgisayar(self):
        for y in range(1,11):
            if self.bilgisayar_random_skor_giris == y:
                if self.kalan_el_sayac == 1:
                    self.rotated_image1 = self.rotate_and_resize("k_"+str(y) + ".png", 0, 0.8)
                    self.photo1 = ImageTk.PhotoImage(self.rotated_image1)
                    self.label1 = tk.Label(self.rakip_frame,bg="#add8e6", image=self.photo1, borderwidth=0, highlightthickness=0,height=0,width=0)
                    self.label1.image = self.photo1
                    self.label1.place(x=70, y=75)
                elif self.kalan_el_sayac == 2:
                    self.rotated_image2 = self.rotate_and_resize("s_"+str(y) + ".png", 0, 0.8)
                    self.photo2 = ImageTk.PhotoImage(self.rotated_image2)
                    self.label2 = tk.Label(self.rakip_frame,bg="#add8e6", image=self.photo2, borderwidth=0, highlightthickness=0,height=0,width=0)
                    self.label2.image = self.photo2
                    self.label2.place(x=90, y=75)
                elif self.kalan_el_sayac == 3:
                    self.rotated_image3 = self.rotate_and_resize("k"+str(y) + ".png", 0, 0.62)
                    self.photo3 = ImageTk.PhotoImage(self.rotated_image3)
                    self.label3 = tk.Label(self.rakip_frame,bg="#add8e6", image=self.photo3, borderwidth=0, highlightthickness=0,height=0,width=0)
                    self.label3.image = self.photo3
                    self.label3.place(x=110, y=75)
                elif self.kalan_el_sayac == 4:
                    self.rotated_image4 = self.rotate_and_resize("s"+str(y) + ".png", 0, 0.55)
                    self.photo4 = ImageTk.PhotoImage(self.rotated_image4)
                    self.label4 = tk.Label(self.rakip_frame,bg="#add8e6", image=self.photo4, borderwidth=0, highlightthickness=0,height=0,width=0)
                    self.label4.image = self.photo4
                    self.label4.place(x=130, y=75)
                else:
                    None
    def bahis_girisi(self):
        self.bahis_miktari = int(app.value)
        if self.bahis_miktari > kredi_miktari.kredim:
            messagebox.showerror("Hata", "Sahip olduğunuz kredi miktarından fazla, bahis miktarı giremezsiniz!!!")
            return self.oyunu_baslat
        self.bahis_text.destroy()
        self.bahis_alani.destroy()
        self.bahis_onay_button.destroy()
        self.girilen_bahis_text.destroy()
        self.girilen_bahis_gorunmesi_text.destroy()
        self.mevcut_kredi_text.destroy()
        self.mevcut_kredi_miktari_text.destroy()
        messagebox.showinfo("Başarılı işlem","Oyununuz başlamıştır. İYİ EĞLENCELER!!!")
        def oyuna_giris():
            self.kalan_el_sayac = 0
            self.oyuncu_random_skor_giris = random.randint(1,10)
            self.bilgisayar_random_skor_giris = random.randint(1,10)
            self.kalan_el_sayac += 1
            self.oyuncunun_skoru += self.oyuncu_random_skor_giris
            self.bilgisayarin_skoru += self.bilgisayar_random_skor_giris
            
            self.cek_button = Button(self.button_frame,
                                     text="Çek",
                                     command=self.cek,
                                     width=8,
                                     height=2,
                                     font="Verdana 12 bold",
                                     bg="#ffffff")
            self.cek_button.place(relx=0.07,rely=0.35)

            self.pas_button = Button(self.button_frame,text="Pas",
                                     command=self.pas,
                                     width=8,
                                     height=2,
                                     font="Verdana 12 bold",
                                     bg="#ffffff")
            self.pas_button.place(relx=0.51,rely=0.35)
            
            self.kalan_hak = Label(self.button_frame,bg="#f0f0f0",text="Kalan Hakkınız: ",font="Verdana 12 bold")
            self.kalan_hak.place(relx=0.17,rely=0.15)

            self.kalan_hak = Label(self.button_frame,bg="#f0f0f0",text=self.oyun_hakki,font="Verdana 12 bold")
            self.kalan_hak.place(relx=0.67,rely=0.155)
            
            self.rakip_skor_text = Label(self.rakip_frame,bg="#add8e6",text=self.bilgisayarin_skoru,font="Verdana 12 bold")
            self.rakip_skor_text.place(relx=0.47,rely=0.7)

            self.oyuncu_skor_text = Label(self.oyuncu_frame,bg="#add8e6",text=self.oyuncunun_skoru,font="Verdana 12 bold")
            self.oyuncu_skor_text.place(relx=0.47,rely=0.7)
            
            self.resim_goruntule_bilgisayar()
            self.resim_goruntule_oyuncu()
        
            
        oyuna_giris()
if __name__ == "__main__":
    master = Tk()
    app = Yirmi_Bir(master)
    master.mainloop()
```
